```r
u='https://raw.githubusercontent.com/DavidQuigley/biostatistics/master/data/metabric.txt'; curtis=read.table(u, header=TRUE, stringsAsFactors=FALSE);
age = curtis$age;
subtype = curtis$er
```

```r
m = mean(age)
```

```r
sd = sd(age)
```

```r
lower_bound = (qnorm(0.05/2, mean = m, sd = sd, lower.tail = TRUE, log.p = FALSE))
upper_bound = (qnorm(0.05/2, mean = m, sd = sd, lower.tail = FALSE, log.p = FALSE))

print(lower_bound)
print(upper_bound)
```

```r
sum( age>lower_bound & age<upper_bound ) / length(age)
```

```r
1-pnorm(87, mean = m, sd = sd)
```
