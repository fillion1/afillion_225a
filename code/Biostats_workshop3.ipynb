```r
u='https://raw.githubusercontent.com/DavidQuigley/biostatistics/master/data/metabric.txt';
curtis=read.table(u, header=TRUE, stringsAsFactors=FALSE);
age = curtis$age;
subtype = curtis$er
```

```r
length(age)
```

```r
length(subtype)
```

```r
hist(age)
```

```r
mean_age = mean(age)
```

```r
median_age = median(age)
```

```r
sd_age = sd(age)
```

```r
upper = mean_age + sd_age
lower = mean_age - sd_age

print(upper)
print(lower)
```

```r
url='https://raw.githubusercontent.com/DavidQuigley/biostatistics/master/data/data_covariates.txt' ;
ds_covar=read.table(url, header=TRUE, stringsAsFactors=FALSE);
assay = ds_covar$bp;
year = ds_covar$year;
sex = ds_covar$sex
```

```r
hist(assay)
```

```r
boxplot(assay ~ sex)
```

```r
boxplot(assay ~ year)
```

```r
layout(matrix(1:2,1,2))
boxplot(assay[year==2007] ~ sex[year==2007], main="2007 results")
boxplot(assay[year==2008] ~ sex[year==2008], main="2008 results")
```

```r
table(sex, year)
```
